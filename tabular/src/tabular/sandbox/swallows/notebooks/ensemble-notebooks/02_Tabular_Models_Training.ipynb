```python
path=Path('data')
```

```python
splits=10
```

```python
def fold_train_func(data, output_model_name):
    p=0.20; lr=1e-3; epochs=30
    learn = tabular_learner(data, layers=[727,353], ps=p, emb_drop=p, metrics=accuracy)
    callbacks = SaveModelCallback(learn, every='improvement', mode='max', monitor='accuracy', name=output_model_name)
    learn.fit_one_cycle(epochs, lr, callbacks=callbacks); learn.recorder.plot_losses()
    return learn
```

```python
train_folds(path, splits, 'data_tab', fold_train_func, 1024)
```
